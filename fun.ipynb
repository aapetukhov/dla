```python
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt
```

```python
duration = 5.0
sr = 22050
t = np.linspace(0, duration, int(duration * sr), False)

frequency = 300.0
sinusoid_wave = 0.5 * np.sin(frequency * t * 2 * np.pi)

linear_signal = t / max(t)

print("Слушаем синусоиду :)")
sd.play(sinusoid_wave, sr)
sd.wait()

# non-monotonic frequency signal:
start_freq, end_freq = 200, 100
nonmono = 0.5 * np.sin((start_freq + (end_freq - start_freq) * t / duration) * t * 2 * np.pi)
sd.play(nonmono, sr)
```

```python
plt.plot(t, linear_signal)
plt.plot(t, sinusoid_wave)
plt.plot(t, nonmono)
```
